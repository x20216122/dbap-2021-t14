```python
def centroid(polygon):
    l = len(polygon)
    x = sum([p[0] for p in polygon])
    y = sum([p[1] for p in polygon])
    return (x/l,y/l)

centroid([[1.1,2.2],[3.1,4.1]])
```

```python
import json


    

polygons = []
with open('world-administrative-boundaries-countries.json') as fp:
    countries = json.load(fp)
    for country in countries:
        coordinates = country['fields']['geo_shape']['coordinates']
        for p0 in coordinates:
            try:
                if (country['fields']['geo_shape']['type'] == 'Polygon'):
                    polygons.append({'c':country['fields']['region_code'], 'p':p0, 'm':centroid(p0)})
                else:
                    for p1 in p0:
                            polygons.append({'c':country['fields']['preferred_term'], 'p':p1, 'm':centroid(p1)})
            except:
                print("Error:", country['fields']['preferred_term'])
    
for p in polygons:
    if p['c']=='Ireland':
        print(p['m'])
        
len(polygons)
```

```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon

for p in polygons:
    point = Point(-66.5807507, -3.3886239)
    polygon = Polygon(p['p'])
    if polygon.contains(point):
        print(p['c'])
```
